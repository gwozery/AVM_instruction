```
import requests as re
import pandas as pd
import json
import io
from fake_useragent import UserAgent
ua = UserAgent()
headers = {'User-Agent': ua.chrome}
url = 'http://aomp.judicial.gov.tw/abbs/opendata/Foreclosure.json'
res = re.get(url, headers=headers)
data = json.loads(res.text)

saledate=[];address=[];saltype=[];budsty=[];checkyn=[];minprice=[];summinprc=[];
dic = {}
for item in data['data'][0:]:
    for land in item['land'][0:]:
        saledate.append(land['saledate'])
        try:
            address.append(land['registeno'])
        except:
            address.append(land['hsimun']+land['ctmd']+land['budadd'])
        saltype.append('land')
        budsty.append(land['area'])
        minprice.append(land['minprice'])
        summinprc.append(land['summinprc'])
        checkyn.append(land['checkyn'])

    for house in item['house'][0:]:
        saledate.append(house['saledate'])
        try:
            address.append(house['registeno'])
        except:
            address.append(house['hsimun']+house['ctmd']+house['budadd'])
        saltype.append('house')
        budsty.append(house['budsty'])
        minprice.append(land['minprice'])
        summinprc.append(land['summinprc'])
        checkyn.append(house['checkyn'])
dic = {
    '拍賣日':saledate,
    '地址':address,
    'Type': saltype,
    '資訊':budsty,
    '拍賣底價':minprice,
    '總拍賣底價':summinprc,
    '是否點交':checkyn
}
df = pd.DataFrame(dic)
df.set_index(["拍賣日"] , inplace=True)
df.sort_values(["拍賣日"], inplace=True)
pd.set_option('display.max_rows', None)
# df.to_csv('法拍屋資訊.csv', encoding = "utf_8_sig") # utf8無BOM 用utf_8_sig
```
